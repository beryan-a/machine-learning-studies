```python
def salary_category(salary):
  if salary < 50000:
    return "low"
  elif 50000<= salary < 80000:
    return "medium"
  else:
    return "high"
```

```python
df["Salary"].apply(salary_category)
```

```python
df["Salary_Category"] = df["Salary"].apply(salary_category) # salary category column ekledik
```

```python
df
```

```python
# istenilen:
# if experience > 10 performance_score + 1
```

```python
def adjust_performance(experience):
  if experience > 10:
    return 1
  else:
    return 0
```

```python
df["Experience"].apply(adjust_performance)
```

```python
df["Adjusted"] = df["Experience"].apply(adjust_performance)
```

```python
df
```

```python
df["New_Score"] = df['Performance_Score'] + df['Adjusted']
```

```python
df
```

```python
#method 2
def adjust_new(row):
  if row["Experience"] > 10:
    return row["Performance_Score"] + 1
  else:
    return row["Performance_Score"]
```

```python
df["Adjusted_Score"] = df.apply(adjust_new , axis=1)
```

```python
df
```

```python
df["Formatted_Name"] = df["Name"].apply(lambda x : x.replace("_", " "))
```
