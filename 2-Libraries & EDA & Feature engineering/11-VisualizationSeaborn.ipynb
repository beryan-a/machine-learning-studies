```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/120-years-of-olympic-history-athletes-and-results/athlete_events.csv')
```

```python
plt.scatter("Height", "Weight", data = df) # datayı dften alacak
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid") # arka plana styling yapar
sns.scatterplot(x = "Height", y = "Weight", data = df)
plt.xlabel("Height of atheletes")
plt.ylabel("Weight of atheletes")
plt.title("atheletes Height vs Weight")
plt.show()
```

```python
sns.set_style("darkgrid") # arka plana styling yapar
sns.scatterplot(x = "Height", y = "Weight", data = df)
plt.xlabel("Height of atheletes")
plt.ylabel("Weight of atheletes")
plt.title("atheletes Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x = "Height", y = "Weight", hue="Sex", data = df) # hue ile cinsiyetleri ayırt etti
plt.xlabel("Height of atheletes")
plt.ylabel("Weight of atheletes")
plt.title("atheletes Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x = "Height", y = "Weight", hue="Medal", data = df) # hue ile madalyaları ayırt etti
plt.xlabel("Height of atheletes")
plt.ylabel("Weight of atheletes")
plt.title("atheletes Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x = "Height", y = "Weight", hue="Sex",style="Medal", size= "Age", data = df) # hue ile cinsiyetleri ayırt etti style ile madalyaları size ile yaşları
plt.xlabel("Height of atheletes")
plt.ylabel("Weight of atheletes")
plt.title("atheletes Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x="Height", y="Weight", hue = "Sex", data=df)
plt.xlabel("Height of atheletes")
plt.ylabel("Weight of atheletes")
plt.title("Atheletes Height vs Weight")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height", hue = "Sex" ,data = df)
plt.ylabel("Number of atheletes")
plt.title("Atheletes Height Distribution")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height", hue = "Sex", kind= "kde" ,data = df) # kde çizgi yapıyor
plt.ylabel("Number of atheletes")
plt.title("Atheletes Height Distribution")
plt.show()
```

```python
sns.barplot(x="Medal", y= "Height", hue="Sex", data=df)
plt.title("Atheletes Height by Medal")
plt.show()
```

```python
sns.catplot(x="Medal", y="Height", hue="Sex", col= "Season", data=df) # kind olarak bar ekleyebilirim
plt.show()
```
