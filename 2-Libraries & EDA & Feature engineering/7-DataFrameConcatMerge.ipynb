# ***MERGE*** (Inner, Outer, Left, Right)

```python
df_merge1 = pd.read_csv('/content/drive/MyDrive/7-merge_data1.csv')
```

```python
df_merge1
```

```python
df_merge2 = pd.read_csv('/content/drive/MyDrive/7-merge_data2.csv')
```

```python
df_merge2  # burada idler karışık üstteki tabloyla bu tabloda kesişen yerler var
# yani burada çok fazla yapılacak şey var kesişen yerleri birleştirmek gibi vs.
```

```python
# merge - outer join
```

```python
df_merge_outer = pd.merge(df_merge1, df_merge2, on='Employee_ID', how='outer')
# on =Employee_ID	 aslında empolyeeid ye göre birleştiriyor
#outer da her şeyi kapsıyor yani herşeyi ile birleştiriyor boş olacak olsa bile
```

```python
df_merge_outer # herşeyi birleştirdi ama boş yerler NaN oldu
```

```python
# merge - left join
```

```python
df_merge_left = pd.merge(df_merge1, df_merge2, on='Employee_ID', how='left')
```

```python
df_merge_left
#sadece birinci tabloyu baz alıyor ikinci tabloda bulunup ilk tabloda bulunmayanları almamış
```

```python
# merge - right join
```

```python
df_merge_right = pd.merge(df_merge1 ,df_merge2, on='Employee_ID', how='right')
```

```python
df_merge_right
# left mergin tam tersi, ikinci tabloyu referans alıyor yani ikinci tabloya ortak olanları ekliyor
```

```python
# merge - inner join
```

```python
df_merge_inner = pd.merge(df_merge1, df_merge2, on='Employee_ID', how='inner')
```

```python
df_merge_inner
# burada sadece iki tablonun kesişim datalarını alıyor yani ortak elemanları alır kalanı atar
```
