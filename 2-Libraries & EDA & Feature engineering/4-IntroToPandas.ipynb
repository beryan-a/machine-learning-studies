# **PANDAS**

The reason of importing NumPy with Pandas:
- Pandas is built on top of NumPy, and many of the operations and data structures in Pandas rely on NumPy arrays. While you can use Pandas without explicitly importing NumPy, it's common practice to import both because you might need NumPy functions or data types when working with Pandas DataFrames or Series.
