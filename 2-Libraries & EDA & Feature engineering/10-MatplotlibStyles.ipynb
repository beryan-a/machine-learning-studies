```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = np.linspace(0,10,20)
```

```python
data1
```

```python
data2 = data1**2
```

```python
plt.subplots()
```

```python
my_fig, my_axes = plt.subplots()
```

```python
type(my_axes)
```

```python
type(my_fig)
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1,data2, color ="#1E0703", alpha = 0.5) #alpha transparan yapıyor
my_axes.plot(data2,data1)
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()
new_axes.plot(data1,data1 + 2, color ="blue", linewidth=2,linestyle=":")
new_axes.plot(data1,data1 + 3, color ="red", linewidth=0.8)
new_axes.plot(data1,data1 + 4, color ="green", linestyle="-.")
new_axes.plot(data1,data1 + 5, color ="yellow", linestyle="--", marker = "o", markersize=8, markerfacecolor="red")# markerfacecolor: markerın içini belirtilen renkle doldurur
new_axes.plot(data1,data1 + 6, color ="#000000", linestyle="--", marker = "+", markersize=4) # marker: veri noktalarını işaretleyen yer
plt.show()
```
