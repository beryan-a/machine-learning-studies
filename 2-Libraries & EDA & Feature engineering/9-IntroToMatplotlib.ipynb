```python
age_list = [10,20,30,30,30,40,50,60,70,75] #listeler aynı boyutta olmalı
```

```python
weight_list = [20,60,80,85,86,87,70,90,95,90] #listeler aynı boyutta olmalı
```

```python
plt.plot(age_list,weight_list, "y") #"y" = yellow, g, b ,r
#başlık, eksen isimleri eklemek için
plt.title("Age vs Weight")
plt.xlabel("Age")
plt.ylabel("Weight")

#grafiği kaydetmek için #(dosyaya soldaki klasörde, istersem indirebilirim)
plt.savefig("age_weight.png")

#göstermek için
plt.show()
```

```python
numpy_arr1 = np.linspace(0,10,20) #sıfır ile on arasında eşit aralıklarla 20 tane sayı verir
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1**3 #numpy_arr1 üç üssü
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1,numpy_arr2, "b*-") # *- , *, --
plt.show()
```

```python
plt.plot(numpy_arr2,numpy_arr1, "g*-") # numpy arrların yerini değiştirdim
plt.show()
```

```python
#ayrı ayrı iki tane grafik görmek istersem
plt.subplot(1,2,1) #1 satır, 2 sütun, 1. grafik (yani iki grafik tek satırda gösterilecek ve iki tane yanyana olduğu için iki kolon)
plt.plot(numpy_arr1,numpy_arr2, "r*-")
plt.subplot(1,2,2) #1 satır, 2 sütün, 2. grafik
plt.plot(numpy_arr2,numpy_arr1, "g--")
plt.show()
#buradaki x eksenleri otomatik olarak kendisi ayarlamış
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.4,0.4]) #bu numaralar grafiğin aralığı ve boyutuyla alakalı
figure_axes.plot(numpy_arr1,numpy_arr2, "r-")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure(dpi = 100) # size

new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])

new_axes.plot(numpy_arr1,numpy_arr1 ** 2, label = "numpy array ** 2")
new_axes.plot(numpy_arr1,numpy_arr1 ** 3, label = "numpy array ** 3")
new_axes.legend(loc=1) # loc = 6,3,... yeri belirtiyor
plt.show()
```
