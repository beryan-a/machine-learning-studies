```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# BOX PLOT
 # median, quartile(25%, 50%, 75%), min max 1.5 * IQR, outlier
```

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])

plt.figure(figsize=(6,5)) # boyut
sns.boxplot( y =data) #boxplot
plt.title('Box Plot')
plt.ylabel("Data Value")
#plt.grid(True) #arka planı ekler
plt.show()
# en yukarıdaki yuvarlak 100 ü simgeler(outlier), en ortadaki çizgide medyanı (31)
```

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,56,56,56,56,56,56,56,56,56,56,56,56,57,60])

plt.figure(figsize=(6,5)) # boyut
sns.boxplot( y =data) #boxplot
plt.title('Box Plot')
plt.ylabel("Data Value")
#plt.grid(True) #arka planı ekler
plt.show()
# burada 100ü çıkardım bu nedenle outlier yok, medyan çok yukarıda
```

```python
df = sns.load_dataset("titanic") # seabornda bulunan titanic datasetini yükledik
```

```python
plt.figure(figsize=(13,4)) #grafikleri biraz büyüttüm

plt.subplot(1,2,1)
sns.boxplot(x="class", y="age", data=df)
plt.xlabel("class")
plt.ylabel("age")
plt.title("age by class")

plt.subplot(1,2,2)
sns.boxplot(x="class", y="fare", data=df)
plt.xlabel("class")
plt.ylabel("fare")
plt.title("fare by class")

plt.tight_layout() # grafiklerin çok sıkışık olduğu durumlarda kullanabiliriz
plt.show()

#çıkarımlar:
#yaş arttıkça sınıf artmış , birinci sınıftakiler daha yaşlı
# ödenmiş para kısmında baya outlierlar var , kendi sınıflarına göre baya fazla ödemişler
```
