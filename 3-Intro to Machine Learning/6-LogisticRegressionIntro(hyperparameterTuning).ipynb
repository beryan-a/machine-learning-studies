```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#bank customers data
df = pd.read_csv('https://raw.githubusercontent.com/atilsamancioglu/MachineLearningNotebooks/refs/heads/main/6-bank_customers.csv')
# data hazır scale edilmiş
```

```python
df.columns
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X=df.drop("subscribed", axis=1)
y=df["subscribed"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
logistic.fit(X_train, y_train) #xtrain ile ytraini kullanarak modeli eğit diyorum
```

```python
y_pred = logistic.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
#accuracy(doğruluk)
#confusion matrix(karmaşıklık matrisi)
#classification report(precision, recall, f score)
```

```python
score = accuracy_score(y_pred, y_test)
print("score: ", score) #0.92 bire yakın yani model çoğunlukla drğru tahminlerde bulunmuş
print("-----------------------------------")
print(classification_report(y_pred, y_test))
print("-----------------------------------")
print("confusion matrix: \n", confusion_matrix(y_pred, y_test))

#doğru buldukları çok yüksek (truePositives(134) ve TrueNegatives(142))
```

# **Hyperparameter Tuning**

```python
model = LogisticRegression()
 # parametreler hakkında daha fazla bilgi için sklearnı ziyaret et
```

```python
penalty = ["l1", "l2", "elasticnet"] # bu parametreleri documantiatioda olduğu gibi yazmam gerekiyor.
c_values = [100, 10, 1, 0.1, 0.01] #cezanın katsayısını belirler yani lambda ile ilişkili
  #cvaluesdaki sayıları kafaya göre belirledi deneme yanılma yoluyla en uygunu seçilir
solver = ["newton-cg", "lbfgs", "liblinear", "sag", "saga", "newton-cholesky"] # bu parametreleri documantiatioda olduğu gibi yazmam gerekiyor.
# buradaki tüm kombinasyonları deneyecek
```

```python
# yukaridakileri sözlüğe çevireceğim
params = dict(penalty=penalty, C=c_values, solver=solver)
```

```python
params
```

```python
#grid search Cv
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator=model, param_grid=params, cv=cv, scoring="accuracy", n_jobs=-1)
# GirdSeacrhcvdeki parametreler zorunlu (ilk ikisi en azından)
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_ # denedikten sonra en iyi parametreleri verdi
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_pred, y_test)
print("score: ", score)
print("-----------------------------------")
print(classification_report(y_pred, y_test))
print("-----------------------------------")
print("confusion matrix: \n", confusion_matrix(y_pred, y_test))
#çok şey değişmemiş, ama zatne data temiz olduğundan ötürü çok fark olmaz
```
