```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/atilsamancioglu/MachineLearningNotebooks/refs/heads/main/7-cyber_attack_data.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
# COLUMN AÇIKLAMALARI
# src_packet_rate -> Source-side packet transmission rate
# dst_packet_rate -> Destination-side packet reception rate
# avg_payload_size -> Average size of payload in packets
# connection_duration -> Duration of the connection (in seconds)
# tcp_flag_count -> Number of TCP flag occurrences
# avg_interarrival_time -> Time between packet arrivals
# failed_login_attempts -> Number of failed login attempts
# unusual_port_activity_score -> Score representing unusual port usage
# session_entropy -> Entropy of session behavior (for anomaly detection)
# avg_response_delay -> Average delay in server response (in ms)
#-------Yukarıdakilere bakarak cevaplamaya çalıştığımız:
      # attack_type -> 0 = Normal, 1 = DDoS, 2 = Port Scan
```

```python
df.info() #data temiz, scale edilmiş, boş yok
```

```python
X = df.drop('attack_type',axis=1)
y = df['attack_type']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
score = accuracy_score(y_pred,y_test)
print("score: ", score)
print("---------------------------------")
print(classification_report(y_pred,y_test))
print("---------------------------------")
print("confusion matrix: \n", confusion_matrix(y_pred,y_test))

#confusion matrixte 82,72,82 bizim için önemli olanlar
```

```python
# buradada aynı şekilde önceki gibi best parametreleri bulabilirz
```

```python
model = LogisticRegression()
```

```python
penalty = ["l1", "l2", "elasticnet"]
c_values = [100, 10, 1, 0.1, 0.01]
solver = ["newton-cg", "lbfgs", "liblinear", "sag", "saga", "newton-cholesky"]
```

```python
params = dict(penalty=penalty, C=c_values, solver=solver)
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator=model, param_grid=params, cv=cv, scoring="accuracy", n_jobs=-1)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_pred,y_test)
print("score: ", score)
print("---------------------------------")
print(classification_report(y_pred,y_test))
print("---------------------------------")
print("confusion matrix: \n", confusion_matrix(y_pred,y_test))

# çok değişiklik yok ama recalllar artmış, confusion matrixte bazı değişiklikler olmuş
```

```python
# ONE vs. REST
```

```python
from sklearn.multiclass import OneVsOneClassifier, OneVsRestClassifier
```

```python
onevsonemodel = OneVsOneClassifier(LogisticRegression()) #içinde estimator isteyecek boş olamaz
```

```python
onevsrestmodel = OneVsRestClassifier(LogisticRegression()) #aynı şekilde içi boş olamaz
```

```python
onevsonemodel.fit(X_train, y_train)
y_pred = onevsonemodel.predict(X_test)
score = accuracy_score(y_pred,y_test)
print("score: ", score)
print("---------------------------------")
print(classification_report(y_pred,y_test))
print("---------------------------------")
print("confusion matrix: \n", confusion_matrix(y_pred,y_test))
#score biraz arttı
#buradaki daha iyi sonuç verdi one vs restten
```

```python
onevsrestmodel.fit(X_train, y_train)
y_pred = onevsrestmodel.predict(X_test)
score = accuracy_score(y_pred,y_test)
print("score: ", score)
print("---------------------------------")
print(classification_report(y_pred,y_test))
print("---------------------------------")
print("confusion matrix: \n", confusion_matrix(y_pred,y_test))
#score biraz azaldı
# önceki one vs one daha iyi sonuç verdi
```
