```python
# Redundancy, multicollinearity(Çoklu doğrusal bağıntı), overfitting
# çok yüksek korelasyonları çıkarmamız mantıklı olacaktır(overfitting oluşturabilir)
```

```python
print(X_train.corr().iloc [0,2])
```

```python
def correlation_for_dropping(df, threshold):
  columns_to_drop = set()
  corr = df.corr()
  for i in range(len(corr.columns)):
    for j in range(i):
       if abs(corr.iloc[i,j])> threshold:
        columns_to_drop.add(corr.columns[i])
  return columns_to_drop
```

```python
columns_dropping = correlation_for_dropping(X_train, 0.85) # tablodaki korelasyonları almış
```

```python
columns_dropping
```

```python
X_train.drop(columns_dropping, axis=1, inplace=True)
X_test.drop(columns_dropping, axis=1, inplace=True)
```

```python
X_train.shape # iki kolon çıktı, korelasyonlarından dolayı (eğitime bir şey katmayacaklar)
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
plt.subplots(figsize=(15,10))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title("X_train")
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title("X_train_scaled")
plt.show()
# iki grafik arasında çok fark var ilki çok dağınık, scaled olan daha iyi
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
linear = LinearRegression()
linear.fit(X_train_scaled, y_train)
y_pred = linear.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Absolute Error: ", mae)
print("Mean Squared Error: ", mse)
print("R2 Score: ", r2) # r2 scroemuz yüksek
plt.scatter(y_test, y_pred)
plt.show()
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(X_train_scaled, y_train)
y_pred = lasso.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Absolute Error: ", mae)
print("Mean Squared Error: ", mse)
print("R2 Score: ", r2) # r2 scoru %95 e düştü diğerleri de artıyor(mse, mae)
plt.scatter(y_test, y_pred)
plt.show()
# belki lasso kullanmamıza çok da gerek yoktur.
```

```python
#ridge ile deneyelim
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(X_train_scaled, y_train)
y_pred = ridge.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Absolute Error: ", mae)
print("Mean Squared Error: ", mse)
print("R2 Score: ", r2) # r2 scoru %98 e yükseldi diğerleri de azaldı(mse, mae)
plt.scatter(y_test, y_pred)
plt.show()
#düz regresyondan çok farkı yok
```

```python
#ElasticNet ile deneyelim
from sklearn.linear_model import ElasticNet
elastic = ElasticNet()
elastic.fit(X_train_scaled, y_train)
y_pred = elastic.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Absolute Error: ", mae)
print("Mean Squared Error: ", mse)
print("R2 Score: ", r2) # r2 scoru %89 a iyice düştü diğerleri de iyice arttı(mse, mae)
plt.scatter(y_test, y_pred)
plt.show()
# belki bu dataya en mantıklısı linear regresyonsur ama gene de cross validation felan kontrol etmemiz gerekiyor
```

***LASSO CROSS VALIDATION***

```python
from sklearn.linear_model import LassoCV #CV cross validation
```

```python
lassocv = LassoCV(cv=5) # parametrelerdeki alpha(ceza katsayısı) aslında lamda ile aynı pythondan dolayı alpha sadece
# lassoCv de n_alphas var -> bu deneyeceği alpha sayısını gösteriyor, değiştirebilirz
lassocv.fit(X_train_scaled, y_train)
y_pred = lassocv.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Absolute Error: ", mae)
print("Mean Squared Error: ", mse)
print("R2 Score: ", r2) # r2 scoru %98 e yükseldi diğerleri de azaldı(mse, mae)
plt.scatter(y_test, y_pred)
plt.show()
```
