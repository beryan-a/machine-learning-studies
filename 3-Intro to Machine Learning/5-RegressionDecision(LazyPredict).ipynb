```python
from lazypredict.Supervised import LazyRegressor
from sklearn import datasets
from sklearn.utils import shuffle
import numpy as np
```

```python
diabetes = datasets.load_diabetes() #sklearnın hazır bir datası
```

```python
type(diabetes) #sklearn objesi olarak geliyor
```

```python
diabetes.DESCR # açıklamasını okuyabiliriz(datanın)
diabetes['DESCR'] # bu yolla da yapabiliriz
```

```python
diabetes.data
```

```python
diabetes.target
```

```python
X, y = shuffle(diabetes.data, diabetes.target, random_state=13)
#X'i dataya veriyor, y'yi Targete veriyor.
```

```python
X = X.astype(np.float32) # X'i float yaptı
```

```python
offset = int(X.shape[0] * 0.9)

#%90'ı tarine atmış %10u teste atmış

X_train, y_train = X[:offset], y[:offset]
X_test, y_test = X[offset:], y[offset:]
```

```python
X_train
```

```python
y_train
```

```python
X[0]
```

```python
y[0]
```

```python
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
```

```python
models, predictions = reg.fit(X_train, X_test, y_train, y_test)
# warninge false dediğimiz için bütün warningleri verecek
```
