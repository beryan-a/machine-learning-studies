INDEPENDENT AND DEPENDENT FEATURES


```python
#independent and dependent features
X = df.iloc[:,:-1] # son kolon hariç hepsini al
y = df.iloc[:,-1] #sadece son kolonu al (bunu yapmak için exam score sonda olmalı zatne)
#yukarıdaki yöntem sadece exam scorun son kolonda bulunduğu müddetçe geçerli

#X = df[['Study Hours', 'Sleep Hours', 'Attendance Rate', 'Social Media Hours']]
#y = df['Exam Score']
```

```python
X.head()
```

```python
y.head()
```

```python
#train - test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=15)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
X_test[0]
```

```python
df.iloc[0]
```

```python
new_student = [[5, 7, 90, 2]]
```

```python
new_student_scaled = scaler.transform(new_student)
```

```python
regression.predict(new_student_scaled) # yaukarıdaki verilere göre öğrenci 90,88... alır
```

```python
new_student2 = [[5, 7, 90, 8]]
```

```python
new_student2_scaled = scaler.transform(new_student2)
```

```python
regression.predict(new_student2_scaled) # yaukarıdaki verilere göre öğrenci 64.934... alır
```

PREDICTION

```python
#prediction
```

```python
y_pred = regression.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print("mse: ", mse)
print("mae: ", mae)
```

```python
score = r2_score(y_test, y_pred)
print("r2 score: ", score)
```

```python
# adjusted score
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
