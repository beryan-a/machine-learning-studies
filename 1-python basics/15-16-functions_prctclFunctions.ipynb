# MAP
- map() fonksiyonu, verilen bir fonksiyonu bir iterable (liste, tuple vb.) üzerindeki her elemana uygular ve bir map nesnesi (iterator) döner. Yani map, sıralama yapmaz; sadece her eleman üzerinde işlem yapar.



```python
help(map)
```

```python
list(map(divideNumber, myList)) #myListteki tüm elemanları divideNumber fonksiyonundan geçirir
```

```python
def controlString(string):
  return "Atil" in string
```

```python
controlString("Atil Samancioglu")
```

```python
controlString("Samancioglu")
```

```python
myStringList = ["Atil", "Atil Samancioglu","Samancioglu","Atlas"]
```

```python
list(map(controlString, myStringList))
```

## FILTER
 - filter: sadece istediğimiz sonuçları filtre etmemize yardımcı olur

```python
list(filter(controlString,myStringList))
```

LEGB -> L->Local, E-> Enclosing, G->Global, B->Built-In







```python
#Global
myString = "Atıl"

def myFunction():
    #Enclosing
    myString = "Atıl 2"
    print(myString)

    def myFunction2():
        #Local
        myString = "Atıl 3"
        print(myString)

    myFunction2()
```

```python
y = 10

def changeY():
    global y #global diye belirttiğimizde globaldeki veriyi değişitrebiliriz
    y = 5
    print(y)
```

```python
changeY()
```
