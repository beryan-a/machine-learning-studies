```python
my_list = [20,30,40,50,60]
```

```python
for num in my_list:
    print(num)
```

```python
for ix in range(len(my_list)): #indexleri verir
    print(ix)
```

```python
for ix in range(len(my_list)):
    print(my_list[ix])
```

```python
#enumerate
```

```python
for element in enumerate(my_list): #indexleri ve değerleri tuple olarak veriyor (index, değer)
  print(element)
```

```python
for (ix,vl) in enumerate(my_list):
  print(vl)
```
