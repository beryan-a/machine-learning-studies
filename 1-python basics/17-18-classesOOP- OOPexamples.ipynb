# Inheritance
- kalıtım almak
- başka bir classtaki özellikleri alıp
  üzerine yeni özellikler eklemek

```python
class Musician():

  def __init__(self, name):
      self.name = name
      print("musician class")

  def test1(self):
      print("test1")

  def test2(self):
      print("test2")
```

```python
atil = Musician("Atil")
```

```python
atil.name
```

```python
atil.test1()
```

```python
atil.test2()
```

```python
class MusicianPlus(Musician): #burada Musician'ı inheritance olarak aldık

    def __init__(self,name):
        Musician.__init__(self,name)
        print("musician plus")

    def test3(self):
        print("test3")

    #override  -> kalıtım aldığım sınıfın üzerine yazmak
    def test1(self):
        print("test1 test1 test1")
```

```python
atlas = MusicianPlus("Atlas")
```

```python
atlas.test1()
```

```python
atlas.test2()
```

```python
atlas.name="Atlas Samancioğlu"
```

```python
type(atlas)
```

```python
atlas.test3()
```

```python
atil.test1()
```

# Polymorphism
- sözlük olarak aynı isimde farklı işlev


```python
class Banana():

  def __init__(self,name):
    self.name = name

  def info(self):
    return f"100 calories {self.name}"

class Apple():

  def __init__(self,name):
    self.name = name

  def info(self):
    return f"150 calories {self.name}"
```

```python
banana = Banana("banana")
```

```python
apple = Apple("apple")
```

```python
banana.name
```

```python
banana.info()
```

```python
apple.info()
```

```python
fruitList = [banana , apple]
```

```python
for fruit in fruitList:
  print(fruit.info()) #same name different work
```

# Encapsulation
- Hapsetmek demek
- oluşturduğumuz değişkenlerin nereden görüntülenebileceği, ulaşılabileceği ile ilgili


```python
class Phone():

  def __init__(self, name, price):
    self.name = name
    self.__price = price #önüne iki tane _ koyarsak sadece bu sınıfın içinden erişebiliriz
    # 51. satırı iki_ koymadan çalıştırdığımızda çalışır

  def info(self):
    print(f"{self.name} price is: {self.__price}") #buraya da __ eklenmeli hata olmaması için

  def changePrice(self,price): #eğer fiyatı değiştirmek isterim diye sadece class içinden bu metodla yapabilirm
    self.__price = price
```

```python
iphone = Phone("iPhone 14", 500)
```

```python
iphone.info()
```

```python
iphone.price = 400 #iPhone price ı 400'e değiştirebildik
#__ ekledikten sonra 400 olmadı
```

```python
iphone.info() #__ ekledikten sonra 400 olmadı
```

```python
iphone.changePrice(1000)
```

```python
iphone.info() #iphone price 1000 olarak değiştirildi
```

# Abstraction
- soyutlama
- soyut sınıf oluşturup, sınıfın kendisini kullanmayıp ama sınıfın yapısını kullanıyoruz.
- sınıfın yapısını kullanırken de kurallarına  uymak zorundayız.

```python
from abc import ABC, abstractmethod
```

```python
class Car(ABC): #inherit alır gibi ABC alıyoruz

  @abstractmethod
  def maxSpeed(self):
    pass
```

```python
#myCar = Car()
```

```python
class Tesla(Car): #car classın yapısını kullandık

  def maxSpeed(self):
    print("200km")
```

```python
tesla = Tesla()
```

```python
tesla.maxSpeed()
```

```python
class Mercedes(Car):

    def maxSpeed(self):
        print("250km")
```

```python
mercedes = Mercedes()
```

```python
mercedes.maxSpeed()
```

# ÖZEL METHODLAR
(python special methods diye araştır)
bir sürü özel metod var. kullanım amacına göre araştırılmalı


```python
class Fruit():

    def __init__(self,name,calories):
        self.name = name
        self.calories = calories

    def __str__(self): #__str__ string
        return f"{self.name}: {self.calories} calories"

    def __len__(self):
        return self.calories
```

```python
myFruit = Fruit("banana",150)
```

```python
myFruit.calories
```

```python
myFruit.name
```

```python
print(myFruit)
```

```python
class Train():

    def __init__(self,name):
        self.name = name

    def __getitem__(self,key): #getitem özel metodlardan biri
        if key == "a":
            return self.name
        else:
            return "Not found"
```

```python
myTrain = Train("myTrain")
```

```python
myTrain["a"]
```
