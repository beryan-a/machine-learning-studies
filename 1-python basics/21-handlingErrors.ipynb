```python
while True:
    try:
        myAge = int(input("enter age: "))
        print(myAge * 2)
        #break #loopun durması için
    except:
        print("enter your age!!!")
    else:
        print("else executed")
        break
    finally: #finally iki türlü çalışacak
        print("finally")
```
