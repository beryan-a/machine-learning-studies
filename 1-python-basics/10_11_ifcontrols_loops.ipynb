```python
my_list = [10,20,30,40,50,60,70]
```

```python
for num in my_list: #num is the variable that hold every single element in the list
    print(num / 5 * 2)
```

```python
print("for loop started")
for x in my_list:
    new_number = x / 5 * 2
    print(new_number)
print("for loop ended")
```

--CONTINUE - BREAK - PASS

```python
my_list = [10,20,30,40,50,60,70]
```

```python
print("for loop started")
for num in my_list:
    print(num)
print("for loop ended")
```

```python
40 in my_list
```

```python
for number in my_list:
    print(number)
    if number == 40:
        print("yes")
        break #Stop checking after the element i was looking for
```

```python
my_list = [10,20,30,40,50,60,70]
```

```python
for number in my_list:
    if number == 40:
        continue #continue ?
    print(number)
```

```python
for num in my_list:
    pass #kodu sonra tanımlamak için
```

# LOOP (While)

```python
#while
```

```python
x=0
```

```python
while x<10:
  print(x)
  x = x+1
```

```python
last_list = [10,20,30,40,50]
```
